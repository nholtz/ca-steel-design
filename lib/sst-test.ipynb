```python
from sst import SST
sst = SST()
```

```python
w =  sst.section('W360x196')
sst.show(w)
```

```python
wl = sst.section_tables(['W','WWF'],'Mass')

def myfunc0(shape,Fy=300.,M=0.):
    My = Fy * shape.Sx / 1E6
    if My >= M:
        return dict(My=My)

s = sst.select_rows(wl,myfunc0,maxn=5,Fy=350.,M=500.)
s[['Mass','Sx','My']]
```

```python
def myfunc(Zx,Fy=350.,Mf=0):
    Mp = Fy * Zx / 1E6
    if Mp >= Mf:
        return dict(Mp=Mp,Foo=13,Fy=Fy,Mf=Mf)
    return False

s = sst.select(wl,myfunc,maxn=10,Mf=2000)
s[['Mass','Avl','Use','Zx','Mp','Foo','Fy','Mf']]
```

```python
s = sst.select(wl,myfunc,Fy=400.,Mf=300.,maxn=20)
s[['Mass','Avl','Use','Zx','Mp','Foo','Fy','Mf']].sort_values('Mp',ascending=False)
```
